## 2. 데이터 분석
"데이터 수집"에서 가져온 데이터를 분석하고, 이 과정을 제시하시오.

2020년 8월까지의 대구 지역의 날씨 데이터와 의복 거래액 데이터 간의 상관관계를 분석한 결과 대구 지역의 날씨 데이터를 통해서 대구의 여름과
겨울의 기온차는 항상 존재해왔다는 것을 알 수 있었고, 의복 거래액 데이터를 통해서 코로나19가 발생하기 이전인 2019년에 비해 코로나19가
발생하고 난 후의 2020년의 의복 거래액 데이터에서 의복 거래액이 2019년 데이터에 비해 확실히 내려갔다는 것을 찾을 수 있었다.

## 3. 시각화
주제에 알맞은 방법으로 시각화 하고, 이를 표현하시오.

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
f = open("./data/daegu.csv", encoding='cp949')
data = csv.reader(f)
summer = []
winter = []
s_date = []
w_date = []
next(data)

for row in data :
    if row[-1] != '' and row[0].split('-')[1] == '06' and row[0].split('-')[2] == '01' :
        s_date.append(float(row[0].split('-')[0]))
        summer.append(float(row[-1]))
    if row[-1] != '' and row[0].split('-')[1] == '12' and row[0].split('-')[2] == '01' :
        w_date.append(float(row[0].split('-')[0]))
        winter.append(float(row[-1]))

plt.figure(figsize=(15,5))
plt.title('Temperature in Daegu # Summer (June), Winter (December)')
plt.plot(s_date,summer,  'mo--', label="Summer")
plt.plot(w_date,winter, 'co--', label="Winter")
plt.grid(True)
plt.xlabel('Years', labelpad=15, fontdict={'family': 'monospace', 'color': 'g', 'weight': 'light', 'size': 20})
plt.ylabel('Temprature', labelpad=20, fontdict={'family': 'monospace', 'color': 'r', 'weight': 'light', 'size': 20})
plt.legend(loc='best', ncol=2)
plt.show()
```
