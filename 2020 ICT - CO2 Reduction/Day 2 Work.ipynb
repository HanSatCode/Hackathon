## 3. 시각화
* 분석 결과를 시각화하시오. 분석 결과 시각화 자료 1 ~ 2개

```python
# The Emissions and Growth volume of carbon dioxide (1946-2018)
import csv
import numpy as np
import matplotlib.pyplot as plt

f = open("./data/ROK_CO2.csv", encoding="utf-8")

data = csv.reader(f)
carbon_E = []
carbon_G = []
carbon_Y = []

next(data) # data의

box1 = {'boxstyle': 'round',
        'ec': (1.0, 0.5, 0.5),
        'fc': (1.0, 0.8, 0.8)}

for row in data :
        carbon_E.append(float(row[-1]))
        carbon_G.append(float(row[-2]))
        carbon_Y.append(float(row[-3]))

plt.figure(figsize=(15,5))
plt.title('The Emissions and Growth volume of carbon dioxide (1946-2018)')
plt.plot(carbon_Y, carbon_E,  'rx-', label="Emissions")
plt.plot(carbon_Y, carbon_G,  'cx--', label="Growth")
plt.grid(True)
plt.xlabel('Years', labelpad=15, fontdict={'family': 'monospace', 'weight': 'light', 'size': 15})
plt.ylabel('Value', labelpad=20, fontdict={'family': 'monospace', 'weight': 'light', 'size': 15})
plt.legend(loc='best', ncol=2)
plt.text(1980, 640, 'It can be seen that the amount of carbon dioxide increases gradually.', rotation=0, bbox=box1)
plt.show()
```

> **우리나라의 이산화탄소 배출량과 증가량에 관한 그래프**다. 이 그래프를 통해 우리나라의 이산화탄소 배출량과 증가량이 점차 증가해나가고 있음을 알 수 있다.

```python
# CO2 Emissions by fuel in South Korea (1945-2018)
import csv
import numpy as np
import matplotlib.pyplot as plt

f = open("./data/ROK_CO2_Fuel.csv", encoding="utf-8")

data = csv.reader(f)
gas = []
coal = []
oil = []
cement = []
year = []

next(data)

for row in data :
        gas.append(float(row[-1]))
        coal.append(float(row[-2]))
        oil.append(float(row[-3]))
        cement.append(float(row[-4]))
        year.append(float(row[-5]))

plt.figure(figsize=(15,5))
plt.title('CO2 Emissions by fuel in South Korea (1945-2018)')

plt.plot(year, gas,  'gx-', label="Gas")
plt.plot(year, coal,  'kx--', label="Coal")
plt.plot(year, oil,  'yx-.', label="Oil")
plt.plot(year, cement,  'cx:', label="Cement")

plt.grid(True)

plt.xlabel('Years', labelpad=15, fontdict={'family': 'monospace', 'weight': 'light', 'size': 15})
plt.ylabel('Value', labelpad=20, fontdict={'family': 'monospace', 'weight': 'light', 'size': 15})
plt.legend(loc='best', ncol=4)

plt.show()
```

> **한국에서 사용하는 연료들을 사용했을 때 발생하는 이산화탄소의 양에 대한 그래프**이다. 이 그래프를 봤을 때 갈수록 석유 에너지를 사용했을 때 발생하는 이산화탄소의 양이 증가하고 있다는 것을 알 수 있다. 그렇기 때문에 여기서 우리는 우리나라의 이산화탄소의 발생량을 감소시키기 위해서 석유 에너지의 사용량을 줄여나가야 한다는 것을 알 수 있다.

## 4. AI 예측
예측을 제시하시오. 예측 시각화 자료 1 ~ 2개 

```python
from fbprophet import Prophet
import pandas as pd
df = pd.read_csv('./data/data2.csv')
m = Prophet(yearly_seasonality=True,weekly_seasonality=False,daily_seasonality=False)
m.fit(df)
future = m.make_future_dataframe(periods=36500)
f_carbon = m.predict(future)
fig= m.plot_components(f_carbon)
```

>다음은 **AI 예측을 통해 만들어낸 한국의 이산화탄소 배출량 증가 예측 그래프**다. 이를 통해 이산화탄소 배출량은 갈수록 증가할 것이라는 것을 알 수 있다. 하지만 여기서 이산화탄소 배출량의 증가는 인간에게 큰 악영향을 끼칠 것이기 때문에 이산화탄소 배출량을 최소한시켜야 한다는 것 역시 알 수 있다.
